```python
import xarray as xr
import numpy as np
```

# Data IO

```python
ds = xr.open_dataset('/maloney-scratch/joedhsu/proj1/data/sst_locking/SPCCSM3.TS.daily.sample.nc',use_cftime=True)
```

# Land asking

```python
from ocean_mask import ocn_mask
 
da_omask_regrid = ocn_mask(ds.TS)
```

```python
ds['TS'] = ds.TS*da_omask_regrid
```

# Calculate background signal

```python
import lanczos_filter as lf
ds['TS_mean'] = ds['TS'].mean(dim='time')
ds['TS_nomean'] = ds['TS']-ds['TS_mean']
ds['TS_bg'] = lf.lanczos_low_pass(ds['TS_nomean'], 201, 1/100., dim='time')+ds['TS_mean']
ds['TS_anom'] = ds['TS']-ds['TS_bg']
```

```python
ds['TS_anom'].i
```

```python
import matplotlib.pyplot as plt

#### plotting
fig=plt.figure(1)
ax1=fig.add_axes([0,0,3,1])
da_ori = ds['TS'].isel(lon=20,lat=10)
da_test = ds['TS_bg'].isel(lon=20,lat=10)
ax1.plot(da_ori.values)
ax1.plot(da_test.values)
ax2=fig.add_axes([0,-1.2,3,1])
da_ori = ds['TS_anom'].isel(lon=20,lat=10)
ax2.plot(da_ori.values)
```

```python
ds=ds.transpose('lat','lon','time')
timeindex = np.arange(len(ds.time)) 
```

# Random Pattern

```python
import random

da_randpatt = ds['TS_anom'].copy()*np.nan
```

```python
%%time
for i in range(1,366):
    dayindex = np.where(ds['time.dayofyear'].values==i)[0]
    random.shuffle(dayindex)
    for newind,oldind in enumerate(dayindex):
        da_randpatt[:,:,newind] = ds['TS_anom'][:,:,oldind] 
```
